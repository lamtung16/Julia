```julia
using Pkg
import("LinearAlgebra")
import("Plots")
```

```julia
function cbf(n,k,t)
  if k==0
    return 1/sqrt(n)
  else
    return sqrt(2/n)*cos(k*(2t+1)*pi/(2n))
  end
end
```

```julia
n=6
C=[cbf(n,i,j) for i=0:(n-1), j=0:(n-1)]
```

```julia
X=[1 .8 1 1 .8 1;1 .5 .3 0 .5 1; 1 .3 .2 0 .3 1; 1 .2 0 0 .2 1; 1 .3 .2 0 .3 1;1  .8  1 1 .8 1]

Y=C*X*C'
```

```julia
using Plots
r=0:(n-1)
M=[r r r r r r]'
xx=reshape(M,(1,36))
yy=reshape(M',(1,36))
zz=reshape(X,(1,36))
scatter(xx,yy,zz, markersize=2,markercolor="red")
```

```julia
using Plots
plotly()

r=0:(n-1)
t=0:.01:(n-1)
P=[cbf(n,k,s) for k=0:(n-1), s in t] 

z=P'*Y*P

plot!(t,t,z,st=:surface)

```

```julia
Mask=[i+j<8 for i=1:6, j=1:6]

Y2=Y.*Mask

z2=P'*Y2*P

scatter(xx,yy,zz, markersize=2,markercolor="red")
plot!(t,t,z2,st=:surface)

#X2=C'*Y2*C
#X-X2
```
